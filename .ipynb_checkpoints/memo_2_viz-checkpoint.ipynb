```python
import pandas as pd
pd.set_option('max_columns', None)
pd.set_option('max_rows', None)
import numpy as np
import plotnine as p9
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data = pd.read_csv('37633-0001-Data.csv')
```

```python
data = data.loc[data['Q16_4'] != -1]
```

```python
print(data.Q16_4.describe())
```

```python
df1 = data[['Q15A_ATGE','Q15A_AHC','Q15A_AHI','Q15A_CE','Q15A_D',
           'Q15A_ER','Q15A_GBM','Q15A_HITH','Q15A_L','Q15A_PCB','Q15A_TBSTLI']]

corrMatrix = df1.corr()
labels = ['Access to\n Education','Access to\n Healthcare','Access to\n Health\n Insurance',
          'Community\n Environment','Discrimination',
           'Economic\n Resources',"Genetics","Access to\n Health Info","Luck",
          "Personal\n Choices","Social\n Biases"]
#corrMatrix = corrMatrix.rename(labels)
plt.figure(figsize=(20,12))
image = sn.heatmap(corrMatrix, annot=True, linewidths = 0.5, xticklabels = labels, yticklabels = labels, 
           center = 0, cmap = 'coolwarm')
image.figure.savefig('heatmap15A.png', dpi = 300)
```
